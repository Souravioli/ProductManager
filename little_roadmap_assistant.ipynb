```python
# %load little_rag.py
# ---------- 2. Imports ----------
import os
import faiss
import langchain_community.vectorstores.faiss as FAISS
import numpy as np
import torch
from pathlib import Path
from sentence_transformers import SentenceTransformer
from transformers import pipeline, set_seed
import json
import requests
import re
from datetime import datetime
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


# ---------- 5. FAISS index ----------
def build_faiss_index(embeddings: np.ndarray) -> faiss.IndexFlatL2:
  """
  Build a FAISS index for fast nearest‑neighbour search.

  Parameters
  ----------
  embeddings : np.ndarray
  2‑D array of document embeddings (float32 or float64).

  Returns
  -------
  faiss.IndexFlatL2
  A FAISS index that can answer distance‑based queries.
  """
  print("Building FAISS index...")
  if embeddings.ndim != 2 or embeddings.shape[0] == 0:
    raise ValueError(
      f"Embeddings must be a non‑empty 2‑D array. Received shape: {embeddings.shape}"
    )
  dim = embeddings.shape[1]
  index = faiss.IndexFlatL2(dim)
  index.add(embeddings)
  return index

# ---------- 6. Retrieval ----------
def retrieve(index: faiss.IndexFlatL2, query_embedding: np.ndarray, k: int = 3) -> tuple:
 """
 Find the top‑k most similar document chunks to a query vector.

 Parameters
 ----------
 index : faiss.IndexFlatL2
 The pre‑built FAISS index.
 query_embedding : np.ndarray
 Embedding of the user question, shape (1, dim).
 k : int, optional
 How many neighbours to return (default 3).

 Returns
 -------
 indices : np.ndarray of int
 1‑D array of the top‑k document indices.
 distances : np.ndarray of float
 Corresponding L2 distances – useful for debugging.
 """
 print(f"Retrieving top-{k} documents...")
 distances, indices = index.search(query_embedding, k)
 return indices[0], distances[0]

# ---------- 7. Prompt construction ----------
def build_prompt(context_docs: list, user_query: str) -> str:
 """
 Assemble the final prompt that will be fed to the language model.

 Parameters
 ----------
 context_docs : list of str
 The text chunks that were retrieved for the query.
 user_query : str
 The original user question.

 Returns
 -------
 prompt : str
 A single string that follows the format required by the
 generation step: “Context:\n<docs>\n\nQuestion: <q>\nAnswer:”
 """
 print("Building prompt...")
 context = "\n\n".join(context_docs)
 prompt = f"Context:\n{context}\n\nQuestion: {user_query}\nAnswer:"
 return prompt

# ---------- 4. Embeddings ----------
def embed_documents(docs: list, model_name: str = "all-MiniLM-L6-v2") -> tuple:
 """
 Convert text chunks into dense vector representations.

 Parameters
 ----------
 docs : list of str
 The list of document chunks to embed.
 model_name : str, optional
 The sentence‑transformer model to use. The default
 “all-MiniLM-L6-v2” is lightweight and works well for quick demos.

 Returns
 -------
 embeddings : np.ndarray
 2‑D array of shape (num_chunks, embedding_dim).
 model : SentenceTransformer
 The loaded embedding model – kept for re‑encoding queries later.
 """
 print("Embedding documents...")
 model = SentenceTransformer(model_name)
 embeddings = model.encode(docs, convert_to_numpy=True)
 return embeddings, model

# ---------- 8. Generation ----------
def generate_answer(context_docs: list, user_query: str) -> str:
 """
 Generate a deterministic answer using a causal language model.

 The generation step is *deterministic* because we fix the random seed
 and set `do_sample=False`. This makes the output reproducible – a
 crucial property for teaching labs.

 Parameters
 ----------
 prompt : str
 Prompt produced by :func:`build_prompt`.
 model_name : str, optional
 Name of the Hugging‑Face transformer model to use.
 `"gpt2-large"` is a good trade‑off between quality and speed.
 max_new_tokens : int, optional
 Maximum number of tokens to generate beyond the prompt.
 dtype : torch.dtype, optional
 Data type for model tensors – `torch.float16` reduces GPU memory
 usage when a GPU is available.

 Returns
 -------
 answer : str
 The generated text after the last “Answer:” marker.
 """
 # Define prompt templates (no need for separate Runnable chains)
 prompt = ChatPromptTemplate.from_template ("Create a response from the context{context} for the question: {question}.")
 context = "\n\n".join(context_docs)

 parser = StrOutputParser()

 # 3. Create the pipeline (chain) using the initialized model
 chain = prompt | llm | parser
 # Run it
 response = chain.invoke(input={"context": context, "question": user_query})
 return response

def load_processed_files(index_file_path):
    """
    Loads a set of processed file paths from a JSON file.
    """
    if index_file_path.exists():
        with open(index_file_path, 'r', encoding='utf-8') as f:
            return set(json.load(f))
    return set()

def save_processed_files(processed_files_set, index_file_path):
    """
    Saves a set of processed file paths to a JSON file.
    """
    with open(index_file_path, 'w', encoding='utf-8') as f:
        json.dump(list(processed_files_set), f, indent=4)

from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/50 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

def find_and_follow(content, output_dir):
        links = re.findall(r'(https?://\S+)', content)

        print(f"Found {len(links)} links. Following...\n\n")

        for link in links:
            # Ensure the output directory exists
            Path(output_dir).mkdir(parents=True, exist_ok=True)
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            # Changed output_filename to be relative to the output_dir
            output_filename = Path(output_dir) / f"output_{timestamp}.txt"

            try:
              with open(output_filename, 'w', encoding='utf-8') as f:
                # Regex to find http/https links
                # Follow the link
                response = requests.get(link, headers=headers, timeout=5)
                if (response.status_code == 200):
                  # Parse HTML and extract text
                  soup = BeautifulSoup(response.text, 'html.parser')
                  plain_text = soup.get_text(separator=' ', strip=True)
                  f.write(f"Content:\n{plain_text}\n\n") # Add this line to write the response text
                else:
                    f.write(f"Failed to retrieve content from {link}. Status Code: {response.status_code}\n\n")
            except Exception as e:
              print(f"Exception {e.__class__.__name__} while processing {link}")
            print(f"Output successfully written to {output_filename}")

# ---------- 3. Document collector ----------
def collect_documents(doc_dir: str):
  """
  Load and preprocess all plain‑text documents in a directory.
  It uses an index to avoid reprocessing already processed files.

  Parameters
  ----------
  doc_dir : str
  Path to a folder that contains one or more *.txt files.

  Returns
  -------
  docs : list of str
  A flat list where each element is a chunk of 50 words taken from the
  original files, with a 10‑word overlap between consecutive chunks.

  Notes
  -----
  • Chunking at a small granularity (50 words) allows the retriever to
  identify highly relevant snippets rather than whole paragraphs.
  • The 10‑word overlap ensures that the boundary words of a chunk
  are not lost when we split a document – this improves semantic
  continuity for the embedding model.
  """
  print("Collecting documents...")
  INDEX_FILENAME = "processed_files_index.json"
  OUTPUT_DIR_FOR_LINKS = Path(doc_dir) / "link_outputs"

  # Ensure the directory for linked outputs and the index file exists
  OUTPUT_DIR_FOR_LINKS.mkdir(parents=True, exist_ok=True)

  index_file_path = OUTPUT_DIR_FOR_LINKS / INDEX_FILENAME
  print(f"index file path {index_file_path}")
  processed_files = load_processed_files(index_file_path)
  print(f"processed files count: {len(processed_files)}")
  files_to_process = []
  for file_path in Path(doc_dir).glob("*.txt"):
    print(f"Checking file: {file_path}")
    if file_path.name == INDEX_FILENAME:
      continue # Skip the index file itself
    if str(file_path) not in processed_files:
      files_to_process.append(file_path)

  if not files_to_process:
    print("No new files to process.")
    return

  for file_path in files_to_process:
    with open(file_path, "r", encoding="utf-8") as f:
      content = f.read()
      print(f"Processing new file: {file_path}")
      find_and_follow(content, OUTPUT_DIR_FOR_LINKS) # Corrected output directory for linked files
      processed_files.add(str(file_path))

  save_processed_files(processed_files, index_file_path)
  print(f"Updated processed files index at {index_file_path}")
  print(f"Processed {len(files_to_process)} new documents.")

# ---------- 3. Document loader ----------
def load_documents(doc_dir: str) -> list:
  INDEX_FILENAME = "processed_files_index.json"
  """
  Load and preprocess all plain‑text documents in a directory.

  Parameters
  ----------
  doc_dir : str
  Path to a folder that contains one or more *.txt files.
  output_link_filename : str, optional
  If provided, links found will be written to this file. Defaults to a timestamped file.

  Returns
  -------
  docs : list of str
  A flat list where each element is a chunk of 50 words taken from the
  original files, with a 10‑word overlap between consecutive chunks.

  Notes
  -----
  • Chunking at a small granularity (50 words) allows the retriever to
  identify highly relevant snippets rather than whole paragraphs.
  • The 10‑word overlap ensures that the boundary words of a chunk
  are not lost when we split a document – this improves semantic
  continuity for the embedding model.
  """
  print("Loading documents...")
  # Helper that splits a single string into overlapping chunks
  def chunk_text(text: str, chunk_size: int = 50, overlap: int = 10) -> list:
    """
    Split a block of text into overlapping word‑based chunks.

    Parameters
    ----------
    text : str
    Raw document text.
    chunk_size : int, optional
    Number of words per chunk (default 50).
    overlap : int, optional
    Number of words that consecutive chunks share (default 10).

    Returns
    -------
    list of str
    List of chunk strings.
    """
    words = text.split()
    chunks = []
    start = 0
    while start < len(words):
      end = start + chunk_size
      chunk = " ".join(words[start:end])
      chunks.append(chunk)
      start += chunk_size - overlap
    return chunks
  docs = []
  #Processing the main files
  for file_path in Path(doc_dir).glob("*.txt"):
    if file_path.name == INDEX_FILENAME:
      continue # Skip the index file itself
    with open(file_path, "r", encoding="utf-8") as f:
      content = f.read()
      # Extend the master list with the new chunks
      docs.extend(chunk_text(content))
  #Processing the linked files
  for file_path in Path(f"{doc_dir}/link_outputs").glob("*.txt"):
    if file_path.name == INDEX_FILENAME:
      continue # Skip the index file itself
    with open(file_path, "r", encoding="utf-8") as f:
      content = f.read()
      # Extend the master list with the new chunks
      docs.extend(chunk_text(content))
  print(f"Loaded: Number of docs: {len(docs)}")
  return docs

from langsmith import traceable

@traceable()
def run_rag_pipeline(user_query: str) -> tuple[str, list]:
  """
  Runs the full RAG pipeline for a given user query.

  Parameters
  ----------
  user_query : str
      The user's question.

  Returns
  -------
  tuple[str, list]
      A tuple containing the generated answer and the list of retrieved documents.
  """

  # 1️⃣ Load documents
  docs = load_documents("/content/drive/MyDrive/Colab Notebooks/docs")
  # 2️⃣ Create embeddings
  #embeddings, embed_model = embed_documents(docs)
  embed_model = SentenceTransformer("all-MiniLM-L6-v2")
  # # 3️⃣ Build FAISS index -try later
  # faiss_index = FAISS.load_local(
  #       "faiss_index",
  #       embeddings,
  #       allow_dangerous_deserialization=True
  #   )
  #faiss_index = faiss.read_index("faiss_index")
  model = SentenceTransformer("all-MiniLM-L6-v2")
  embeddings = model.encode(docs, convert_to_numpy=True)
  faiss_index = build_faiss_index(embeddings)
  # 4️⃣ Encode the user query
  query_vec = embed_model.encode([user_query], convert_to_numpy=True)

  # 5️⃣ Retrieve top‑k contexts
  top_k_indices, _ = retrieve(faiss_index, query_vec, k=3)
  retrieved_docs = [docs[i] for i in top_k_indices]

  # 6️⃣ Build prompt
  #prompt = build_prompt(retrieved_docs, user_query)
  #print (f"prompt {prompt}")
  # 7️⃣ Generate answer
  answer = generate_answer(
      context_docs=retrieved_docs,
      user_query=user_query
  )
  return answer, retrieved_docs

from typing import Annotated, TypedDict
from langchain_openai import ChatOpenAI # Assuming ChatOpenAI is used for retrieval_relevance_llm

# Grade output schema
class RetrievalRelevanceGrade(TypedDict):
    explanation: Annotated[str, ..., "Explain your reasoning for the score"]
    score: Annotated[
        int,
        ...,
        "A relevance score between 1 and 5. 1 if none or little relevant information, 2-5 for increasing relevance.",
    ]

# Grade prompt
retrieval_relevance_instructions = """You are a analyist grading an answer. You will be given a Product Market QUESTION and a set of FACTS provided by the researcher. Your goal is to assign a relevance score to the FACTS based on the QUESTION.

Scoring Criteria:
- Score 1-2: FACTS contain very little or vaguely related information, or contain some keywords but lack semantic meaning.
- Score 3-4: FACTS contain some relevant information, addressing parts of the question, or have a clear semantic connection.
- Score 5: FACTS are highly relevant and directly answer the QUESTION comprehensively.

Explain your reasoning in a step-by-step manner to ensure your reasoning and conclusion are correct. Avoid simply stating the correct answer at the outset."""

# Grader LLM
retrieval_relevance_llm = ChatOpenAI(
    model="gpt-4o-mini", temperature=0 # Changed model to gpt-4o-mini as used elsewhere
).with_structured_output(RetrievalRelevanceGrade, method="json_schema", strict=True)

print("RetrievalRelevanceGrade and retrieval_relevance_llm updated to provide a numeric score.")

def retrieval_relevance(inputs: dict, outputs: dict) -> int:
    """An evaluator for document relevance, returning a numeric score from 0-5."""
    # Ensure outputs['documents'] is a list of strings if it contains page_content, otherwise handle it
    if outputs and 'documents' in outputs and outputs['documents']:
        if hasattr(outputs['documents'][0], 'page_content'): # Check if it's a list of Document objects
            doc_string = "\n\n".join(doc.page_content for doc in outputs["documents"])
        else: # Assume it's already a list of strings
            doc_string = "\n\n".join(outputs["documents"])
    else:
        doc_string = "No documents retrieved."

    answer = f"FACTS: {doc_string}\nQUESTION: {inputs['question']}"
    # Run evaluator
    grade = retrieval_relevance_llm.invoke([
        {"role": "system", "content": retrieval_relevance_instructions},
        {"role": "user", "content": answer}
    ])
    return grade["score"]

from langsmith import traceable

@traceable()
def rag_bot(user_query: str) -> dict:
    # langchain Retriever will be automatically traced
  answer, retrieved_docs = run_rag_pipeline(user_query) # user_query is a string here

  return {"outputs": answer, "documents": retrieved_docs}

def target(inputs: dict) -> dict:
  # Ensure inputs["question"] is a string before passing to rag_bot
  question_input = inputs["question"]
  print(f"question_input in target: {question_input}")
  actual_question_str = ""
  if isinstance(question_input, list) and len(question_input) > 0:
      actual_question_str = question_input[0]
  elif isinstance(question_input, str):
      actual_question_str = question_input
  else:
      raise ValueError(f"Question input has unexpected type: {type(question_input)}. Expected str or list[str].")

  return rag_bot(actual_question_str) # Pass the string to rag_bot

# class FeatureImportanceInput(TypedDict):
#     competitor_name: Annotated[str, ..., "The name of the competitor to be evaluated"]
#     features: Annotated[list[str], ..., "A list of features or criteria to analyze"]

# class FeatureImportanceOutput(TypedDict):
#     analysis_summary: Annotated[str, ..., "The synthesized analysis output from the LLM"]
#     feature_scores: Annotated[dict[str, int], ..., "Mapping of features to their evaluated scores (1-5)"]

# class CompetitorState(TypedDict):
#     feature_importance_input: FeatureImportanceInput
#     feature_importance_output: FeatureImportanceOutput
#     retrieved_docs: Annotated[list[str], ..., "The document chunks used for this analysis"]

# @traceable()
# def query_competition_agent(state: CompetitorState) -> dict:
#     """
#     Agent function to perform competition analysis using the RAG pipeline.
#     """
#     input_data = state.get("feature_importance_input")
#     if not input_data:
#         raise ValueError("feature_importance_input is missing from state")

#     comp_name = input_data.get("competitor_name", "Unknown Competitor")
#     features = input_data.get("features", [])

#     query = f"Analyze {comp_name} focusing on: {', '.join(features)}"

#     # Utilize the existing RAG pipeline logic
#     answer, docs = run_rag_pipeline(query)

#     return {
#         "feature_importance_output": {"analysis_summary": answer, "feature_scores": {}},
#         "retrieved_docs": docs
#     }
```

```python
from pydantic import BaseModel, Field
from typing import List, Literal, Optional, Union

class ResourceRequired(BaseModel):
    skill_set: str = Field(description="The skill set required, e.g., 'Senior Frontend developer', 'QA'.")
    num_resources: int = Field(description="Number of resources with this skill set.")
    man_months: int = Field(description="Total man-months of effort required for this skill set for this feature.")

class FeatureItem(BaseModel):
    feature_name: str = Field(description="The name of the feature.")
    functionality:str = Field(description="The functionality this feature is associated with, or if the feature itself is a functionality. E.g., 'Get Application Inventory'.")
    resources_required: List[ResourceRequired] = Field(description="List of resources and effort required for the feature.")
    mandatory: bool = Field(description="True if the feature is mandatory, False otherwise.")

class FeatureValue(BaseModel):
    """Represents the evaluated value of a feature."""
    feature_name: str = Field(description="The name of the evaluated feature.")
    overall_value_score: float = Field(description="The calculated overall value score for the feature.")
    value_explanation: str = Field(description="A detailed explanation of how the overall value score was derived.")

class FeatureImportanceInput(BaseModel):
    """Input schema for evaluating feature importance in a specific market."""
    market_name: str = Field(description="The name of the market, e.g., 'E-commerce', 'Exposure Management'.")
    feature_name: str = Field(description="The name of the feature, e.g., 'AI-powered Recommendation Engine'.")
    functionality_name: str = Field(description="The specific functionality of the feature, e.g., 'Enhance User Experience', 'Attack Path Discovery'.")
class FeatureImportanceOutput(BaseModel):
    """Output schema for the evaluated feature importance."""
    feature_value: List[FeatureValue] = Field(description="A list of individual feature importance values and their explanations.")
    #importance_score: int = Field(description="An importance score between 1 (very low) and 5 (very high) for the feature and functionality in the given market.")
    #reasoning: str = Field(description="A detailed explanation of how the importance score was derived, based on retrieved information.")

print("Pydantic models for Feature Importance defined.")

class Feature(BaseModel):
    """Represents a feature to be evaluated."""
    name: str = Field(description="The name of the feature.")
    functionality: str = Field(description="The functionality this feature is associated with, or if the feature itself is a functionality. E.g., 'Get Application Inventory'.")
    feature_type: Literal["Usability", "Performance", "General", "Vulnerability", "Component refresh", "Technical debt"] = Field(description="The type of the feature, influencing its value based on market stage.")
    priority: Literal["low", "medium", "high", "critical"] = Field(description="The priority of the feature as provided by a customer.")
    feature_category: Literal["innovation", "differentiation"] = Field(description="Whether the feature is an innovation or a differentiation.")
    innovation_score: int = Field(description="Innovation score on a scale of 1 to 10. Only applicable if 'feature_category' is 'innovation'. Differentiation features get a default score of 4.") # Changed to mandatory int
    targets_new_user_persona: bool = Field(description="True if the functionality is meant to target a new user persona, False otherwise.")
    targets_new_market: bool = Field(description="True if the functionality is meant to target a new market, False otherwise.")
    current_market: str = Field(description="The current market for the product, e.g., 'Exposure Management'.")
    market_stage: Literal["new", "growth", "mature"] = Field(description="The current stage of the product's market.")


# New Pydantic model for inputting a list of features
class FeaturesInputList(BaseModel):
    features: List[Feature] = Field(description="A list of Feature objects to evaluate.")

# New Pydantic model for aggregated output
class TotalFeatureValueOutput(BaseModel):
    """Output schema for the aggregated feature value of multiple features."""
    total_value_score: float = Field(description="The sum of overall value scores for all features provided.")
    individual_feature_values: List[FeatureValue] = Field(description="A list of individual feature values and their explanations.")
    reasoning: str = Field(description="A comprehensive reasoning explaining the calculation of the total feature value.")

class PerceivedValueInput(BaseModel):
    """Input schema for the Perceived Value Agent."""
    features: List[Feature] = Field(description="A list of features to evaluate.")
    supplied_marketing_budget: float = Field(description="The initial marketing budget in USD (e.g., 1000000 for $1M).")
    current_market: str = Field(description="The current market in which the product operates (e.g., 'Exposure Management').")
    market_stage: Literal["new", "growth", "mature"] = Field(description="The current stage of the product's market.")

class PerceivedValueOutput(BaseModel):
    """Output schema for the Perceived Value Agent."""
    total_perceived_value: float = Field(description="The total calculated perceived value.")
    adjusted_marketing_budget_usd: float = Field(description="The marketing budget after deductions for new persona/market targeting.")
    perceived_value_from_marketing: float = Field(description="The perceived value attributed to the marketing budget.")
    perceived_value_from_features: float = Field(description="The perceived value attributed to innovative/differentiated features.")
    perceived_value_from_target_markets: float = Field(description="The perceived value attributed to targeting new user personas/markets.")
    reasoning: str = Field(description="A detailed explanation of the perceived value computation, including all steps and values.")

print("Pydantic models for Perceived Value Agent and Feature Value Agent updated.")
```

```python

from pydantic import BaseModel, Field
from typing import List, Dict, Any, Optional, Literal
from langchain_core.tools import Tool
from langchain_core.tools import StructuredTool
from langchain_core.messages import SystemMessage
from langchain.agents import create_agent
from langchain.agents.structured_output import ProviderStrategy

import json
import copy

# --- 1. Define Pydantic models for Resource Manager Agent ---
class ResourceManagerInput(BaseModel):
    featureset1: List[FeatureItem] = Field(description="List of features to be evaluated.")
    total_budget: float = Field(description="The total available budget in dollars.")
    total_time_months: int = Field(description="The total time period in months within which all features must be completed.")
    market_stage: Literal["new", "growth", "mature"] = Field(description="The current stage of the product's market.")

class SelectedFeature(BaseModel):
    feature_name: str = Field(description="The name of the scheduled feature.")
    cost: float = Field(description="The estimated cost of this feature.")
    value: Optional[float] = Field(None, description="The total calculated value.")

class ResourceManagerOutput(BaseModel):
    feasible: bool = Field(description="True if all selected features can be completed within the budget and time, False otherwise.")
    reasoning: str = Field(description="Detailed explanation of the feasibility assessment, including cost breakdown and time considerations.")
    total_estimated_cost: Optional[float] = Field(None, description="The total estimated cost for all features, including reservations, if feasible.")
    total_estimated_development_months: Optional[int] = Field(None, description="The total estimated development duration in months (sequential), if feasible.")
    selected_features: Optional[List[SelectedFeature]] = Field(None, description="List of selected features for the roadmap.")

print("Pydantic models for Resource Manager Agent defined.")

# --- 2. Define the core logic function 'evaluate_feature_completion_function' ---
SKILL_MONTHLY_COSTS = {
    "Senior Frontend developer": 5000,
    "Junior Frontend Developer": 4000,
    "Senior Backend Developer": 6000,
    "Junior Backend Developer": 4500,
    "QA": 4000,
    "Other Resource": 5000,
}

def evaluate_feature_completion_function(
    featureset1: List[FeatureItem],
    total_budget: float,
    total_time_months: int,
    market_stage: Literal["new", "growth", "mature"]
) -> ResourceManagerOutput:
    reasoning_steps = []

    if not featureset1:
        return ResourceManagerOutput(
            feasible=False,
            reasoning="No features provided to evaluate.",
            total_estimated_cost=0.0,
            total_estimated_development_months=0,
            selected_features=[]
        )
    print(f"Supplied budget for feature = {total_budget}")

    def calculate_metrics_for_set(features_subset: List[FeatureItem]):
        current_raw_development_cost = 0.0
        current_required_man_months_per_skill = {skill: 0.0 for skill in SKILL_MONTHLY_COSTS.keys()}
        current_duration_man_months_per_skill = {skill: 0.0 for skill in SKILL_MONTHLY_COSTS.keys()}
        for feature in features_subset:
            feature_current_raw_cost = 0.0
            feature_man_months_for_other_resources = 0.0
            feature_duration_for_other_resources = 0.0
            has_explicit_qa = any(r.skill_set == "QA" for r in feature.resources_required)

            for resource in feature.resources_required:
                cost_per_resource_type = SKILL_MONTHLY_COSTS.get(resource.skill_set, SKILL_MONTHLY_COSTS["Other Resource"])
                resource_total_effort_cost = resource.num_resources * resource.man_months * cost_per_resource_type
                feature_current_raw_cost += resource_total_effort_cost
                current_required_man_months_per_skill[resource.skill_set] += (resource.num_resources * resource.man_months)
                current_duration_man_months_per_skill[resource.skill_set] += (resource.man_months)
                if resource.skill_set != "QA":
                    feature_duration_for_other_resources += resource.man_months
                    feature_man_months_for_other_resources += (resource.num_resources*resource.man_months)

            if not has_explicit_qa:
                qa_duration_months = 0.5 * feature_duration_for_other_resources
                qa_cost_man_months = 0.5 * feature_man_months_for_other_resources
                qa_man_months = qa_duration_months + qa_cost_man_months
                implicit_qa_cost = qa_cost_man_months * SKILL_MONTHLY_COSTS["QA"]
                feature_current_raw_cost += implicit_qa_cost
                current_duration_man_months_per_skill["QA"] += qa_duration_months

            current_raw_development_cost += feature_current_raw_cost

        bug_fix_reservation_rate = 0.3 if market_stage == "new" else 0.2
        management_overhead_rate = 0.2
        total_reservation_factor = 1 + bug_fix_reservation_rate + management_overhead_rate
        current_estimated_cost_with_reservations = current_raw_development_cost * total_reservation_factor

        return current_estimated_cost_with_reservations, current_duration_man_months_per_skill

    # Feature Selection
    mandatory_features = [f for f in featureset1 if f.mandatory]
    optional_features = [f for f in featureset1 if not f.mandatory]

    # Check Mandatory first
    current_selected = list(mandatory_features)
    reasoning_steps.append(f"Processing {len(mandatory_features)} mandatory features.")

    est_cost, mm_per_skill = calculate_metrics_for_set(current_selected)

    # Calculate max development months for mandatory features
    max_mandatory_dev_months = 0
    if mm_per_skill: # Ensure mm_per_skill is not empty
        max_mandatory_dev_months = max(mm_per_skill.values())

    # Check constraints for mandatory
    mandatory_feasible = est_cost <= total_budget and all(mm <= total_time_months for mm in mm_per_skill.values())

    if not mandatory_feasible:
        reasoning_steps.append("Mandatory features alone exceed budget or time constraints.")
        # Recalculate cost for individual mandatory features for the output
        selected_features_output = []
        for f in current_selected:
            individual_cost, _ = calculate_metrics_for_set([f])
            selected_features_output.append(SelectedFeature(feature_name=f.feature_name, cost=individual_cost))
        return ResourceManagerOutput(feasible=False, reasoning="\n".join(reasoning_steps), selected_features=selected_features_output)

    # Sort optional by cost (cheapest first) to maximize count
    optional_with_costs = []
    for f in optional_features:
        cost, _ = calculate_metrics_for_set([f])
        optional_with_costs.append((cost, f))
    optional_with_costs.sort(key=lambda x: x[0])

    # Initialize max_development_months with the value from mandatory features
    max_development_months = max_mandatory_dev_months

    for cost, feat in optional_with_costs:
        tentative = current_selected + [feat]
        t_cost, t_mm = calculate_metrics_for_set(tentative)

        # Update max_development_months if this feature increases it
        current_max_tentative_mm = max(t_mm.values()) if t_mm else 0

        if t_cost <= total_budget and all(mm <= total_time_months for mm in t_mm.values()):
            current_selected.append(feat)
            est_cost = t_cost
            max_development_months = max(max_development_months, current_max_tentative_mm)
            reasoning_steps.append(f"Added optional feature '{feat.feature_name}' to maximize total features.")

    reasoning_steps.append(f"Final selection contains {len(current_selected)} features with an estimated cost of ${est_cost:,.2f}.")
    print(f"Total months for this cycle: {max_development_months}")

    selected_features_output = []
    for f in current_selected:
        individual_cost, _ = calculate_metrics_for_set([f]) # Recalculate individual cost for output
        selected_features_output.append(SelectedFeature(feature_name=f.feature_name, cost=individual_cost))
        print(f"Selected feature '{f.feature_name}' with cost ${individual_cost:,.2f}.")

    return ResourceManagerOutput(
        feasible=True,
        reasoning="\n".join(reasoning_steps),
        total_estimated_cost=est_cost,
        total_estimated_development_months=int(max_development_months), # Use the calculated max development months
        selected_features=selected_features_output
    )

resource_manager_tool = StructuredTool.from_function(
    name="evaluate_feature_completion",
    func=evaluate_feature_completion_function,
    args_schema=ResourceManagerInput,
    description="Evaluates roadmap feasibility."
)

resource_manager_agent = create_agent(
    model=llm,
    name="ResourceManagerAgent",
    tools=[resource_manager_tool],
    response_format=ProviderStrategy(ResourceManagerOutput),
    system_prompt=SystemMessage(content="You are a resource manager.")
)

print("Resource Manager Agent updated to maximize feature count.")
```

```python
# Assuming 'llm' is already defined from previous steps
#llm = ChatOpenAI(temperature=0.7, model_name='gpt-4o-mini') # Re-initialize llm for safety if this cell is run independently

def evaluate_single_feature_value_function(
    feature: Feature
) -> FeatureValue:
    """
    Evaluates the value of a given feature based on predefined criteria.
    This function encapsulates the logic for evaluating individual features.
    """
    base_score = 0.0
    explanation_steps = []

    # 1. Initial Score from Innovation/Differentiation
    current_innovation_score = 0
    if feature.feature_category == "innovation":
        current_innovation_score = feature.innovation_score if feature.innovation_score is not None else 5 # Default for innovation if not provided
        explanation_steps.append(f"Feature is an innovation with a score of {current_innovation_score}.")
    elif feature.feature_category == "differentiation":
        current_innovation_score = 4
        explanation_steps.append(f"Feature is a differentiation, assigned a default innovation score of {current_innovation_score}.")
    base_score += current_innovation_score

    # 2. Add Feature Type Score based on Market Stage
    feature_type_score_for_market = FEATURE_TYPE_SCORES.get(feature.market_stage, {}).get(feature.feature_type.lower())
    if feature_type_score_for_market is not None:
        base_score += feature_type_score_for_market
        explanation_steps.append(f"Added {feature_type_score_for_market} points for feature type '{feature.feature_type}' in a '{feature.market_stage}' market.")
    else:
        explanation_steps.append(f"No specific score found for feature type '{feature.feature_type}' in a '{feature.market_stage}' market.")

    # 3. Add Customer Priority Score (additive based on market stage)
    effective_priority = feature.priority
    if effective_priority is None: # If priority is not defined in the input, try fetching from JIRA
        jira_priority_score = search_jira_for_priority(feature.name)
        if jira_priority_score is not None:
            # Map numerical JIRA score (1-10) to literal categories
            if 1 <= jira_priority_score <= 3:
                effective_priority = "low"
            elif 4 <= jira_priority_score <= 6:
                effective_priority = "medium"
            elif 7 <= jira_priority_score <= 9:
                effective_priority = "high"
            elif jira_priority_score == 10:
                effective_priority = "critical"
            explanation_steps.append(f"Customer priority for '{feature.name}' fetched from JIRA: {effective_priority} (score: {jira_priority_score}).")
        else:
            explanation_steps.append(f"Customer priority for '{feature.name}' not provided and not found in JIRA. Skipping priority score addition.")

    if effective_priority:
        priority_additive = CUSTOMER_PRIORITY_ADDITIVE_SCORE.get(effective_priority)
        if priority_additive is not None:
            base_score += priority_additive
            explanation_steps.append(f"Added {priority_additive} points for customer priority '{effective_priority}' in a '{feature.market_stage}' market.")
        else:
            explanation_steps.append(f"No specific additive score found for customer priority '{effective_priority}' in a '{feature.market_stage}' market.")

    # 4. Core Functionality Multiplier (40% more value)
    if feature.functionality.lower() in CORE_FUNCTIONALITIES_BASE_VALUE:
        base_score *= 1.4
        explanation_steps.append(f"Functionality '{feature.functionality}' is a core functionality, applying a 40% value increase.")
    else:
        explanation_steps.append(f"Functionality '{feature.functionality}' is not a core functionality.")

    # 5. Targeting new user persona
    if feature.targets_new_user_persona:
        base_score += 3  # Add points for targeting new persona
        explanation_steps.append("Targets a new user persona, adding 3 points.")

    # 6. Targeting new market
    if feature.targets_new_market:
        base_score += 5  # Add points for targeting new market
        explanation_steps.append("Targets a new market, adding 5 points.")

    overall_value_score = round(base_score, 2)*4   #multiplier of 10 arbitrary added for now
    value_explanation = "\n".join(explanation_steps)

    return FeatureValue(
        feature_name=feature.name,
        overall_value_score=overall_value_score,
        value_explanation=value_explanation
    )

print("Function 'evaluate_single_feature_value_function' defined with detailed logic.")

def evaluate_multi_feature_value_function(features: List[Feature]) -> TotalFeatureValueOutput:
    """
    Evaluates the combined value of a list of features.
    """
    individual_values: List[FeatureValue] = []
    total_score = 0.0
    overall_explanation_steps = []

    for feature in features:
        feature_value = evaluate_single_feature_value_function(feature)
        individual_values.append(feature_value)
        total_score += feature_value.overall_value_score
        overall_explanation_steps.append(f"- {feature_value.feature_name}: {feature_value.overall_value_score:.2f} ({{feature_value.value_explanation}})")

    overall_explanation = "Combined Feature Value Report:\n" + "\n".join(overall_explanation_steps)
    overall_explanation += f"\n\nTotal Overall Value Score: {total_score:.2f}"

    return TotalFeatureValueOutput(
        individual_feature_values=individual_values,
        total_value_score=round(total_score, 2),
        reasoning=overall_explanation
    )

print("Function 'evaluate_multi_feature_value_function' defined.")

# Create a StructuredTool from the function
multi_feature_value_evaluator_tool = StructuredTool.from_function(
    name="evaluate_multi_feature_value",
    func=evaluate_multi_feature_value_function,
    args_schema=FeaturesInputList, # Use the new Pydantic input model
    description="""Useful for evaluating the combined value of multiple features. This tool calculates an individual value for each feature and then sums them up to provide a total overall value score.
    Input should be a JSON object adhering to the 'FeaturesInputList' Pydantic model schema, containing a list of 'Feature' objects.
    Example input:
    {
        "features": [
            {
                "name": "AI-powered Recommendation Engine",
                "functionality": "Enhance User Experience",
                "feature_type": "Usability",
                "priority": "high",
                "feature_category": "innovation",
                "innovation_score": 8,
                "targets_new_user_persona": true,
                "targets_new_market": false,
                "current_market": "E-commerce",
                "market_stage": "growth"
            },
            {
                "name": "Improved Checkout Flow",
                "functionality": "Streamline Purchase Process",
                "feature_type": "Performance",
                "priority": "critical",
                "feature_category": "differentiation",
                "innovation_score": 4,
                "targets_new_user_persona": false,
                "targets_new_market": false,
                "current_market": "E-commerce",
                "market_stage": "growth"
            }
        ]
    }
    The tool returns a 'TotalFeatureValueOutput' object, including a list of 'FeatureValue' for each feature, a 'total_value_score', and a comprehensive 'reasoning'."""
)

# Define the system message for the FeatureValueAgent
feature_value_system_message = SystemMessage(content=(
    "You are a FeatureValueAgent. Your role is to assess the overall value of a list of product features "
    "based on their characteristics, market context, and strategic impact. You have access to the 'evaluate_multi_feature_value' tool. "
    "When a user provides details about multiple features, use this tool to calculate "
    "and explain the total overall value. Always return a comprehensive TotalFeatureValueOutput object, "
    "including individual feature evaluations and detailed reasoning." +
    "\n\nIMPORTANT: When calling the 'evaluate_multi_feature_value' tool, you must pass the entire FeaturesInputList as a single JSON object argument to the tool."
))

# Create the FeatureValueAgent
feature_value_agent = create_agent(
    model=llm,
    name="FeatureValueAgent",
    tools=[multi_feature_value_evaluator_tool],
    response_format=ProviderStrategy(TotalFeatureValueOutput),
    system_prompt=feature_value_system_message,
)

print("LangChain Tool 'multi_feature_value_evaluator_tool' and 'FeatureValueAgent' created.")
```

```python
from typing import Annotated, Literal, TypedDict, List, Optional, Dict, Union
import operator
from pydantic import BaseModel, Field

# Define the input structure for the entire workflow
class RoadmapInput(BaseModel):
    features: List[Feature] = Field(description="A list of Feature objects for the roadmap.")
    resources_per_feature: Dict[str, List[ResourceRequired]] = Field(description="Dictionary mapping feature names to their resource requirements.")
    mandatory_features: List[str] = Field(description="List of names of mandatory features.")
    total_budget: float = Field(description="The total available budget in dollars for the roadmap.")
    total_time_months: int = Field(description="The total time period in months within which all features must be completed.")
    supplied_marketing_budget: float = Field(description="The initial marketing budget in USD.")
    current_market: str = Field(description="The current market for the product.")
    market_stage: Literal["new", "growth", "mature"] = Field(description="The current stage of the product's market.")

# Define the output structure for the entire workflow
class RoadmapOutput(BaseModel):
    feasible: bool = Field(description="True if the roadmap is feasible within budget and time, False otherwise.")
    total_value: float = Field(description="Total value computed for roadmap")
    Final_marketing_budget_usd: float = Field(description="The final marketing budget in USD.")
    resource_manager_output: Optional[ResourceManagerOutput] = Field(None, description="Detailed output from the Resource Manager Agent.")
    total_feature_value_output: Optional[TotalFeatureValueOutput] = Field(None, description="Detailed output from the Feature Value Agent.")
    selected_features: Optional[List[SelectedFeature]] = Field(description="List of selected feature names, their cost and their value.")
    perceived_value_output: Optional[PerceivedValueOutput] = Field(None, description="Detailed output from the Perceived Value Agent.")
    summary_reasoning: str = Field(description="A comprehensive summary of the roadmap evaluation.")

# Define the state for the LangGraph workflow
class RouterState(TypedDict):
    decision:Optional[str]
    iteration: int # Make it int and mandatory, starting at 0
    roadmap_input: RoadmapInput
    resource_manager_input: Optional[ResourceManagerInput]
    feature_value_input: Optional[FeaturesInputList]
    perceived_value_input: Optional[PerceivedValueInput]
    resource_manager_raw_output: Optional[str]
    resource_manager_output:Optional[ResourceManagerOutput]
    feature_value_raw_output: Optional[str]
    perceived_value_raw_output: Optional[str]
    intermediate_roadmap_output: Optional[RoadmapOutput]
    final_roadmap_output: Optional[RoadmapOutput]
    prev_total_value: Optional[float]
    prev_decision: Optional[str] # To track the decision that led to prev_total_value
    error_reasons:Optional[str] # To track intermediate errors
    all_features_importance: Optional[List[FeatureValue]] # New field to store all feature importance

class CompetitionState(TypedDict):
    feature_list: List[Feature] # Make it mandatory
    current_market:str #current market for product, mandatory
    feature_importance_raw_output: Optional[str]
    feature_importance_output:Optional[List[FeatureValue]]
    error_reasons:Optional[str] # To track intermediate errors
    index:int #Make it mandatory
```

```python
from langgraph.graph import StateGraph, START, END
from langgraph.types import Send
import json
from langchain_core.messages import HumanMessage

# Node to prepare specific inputs for each agent from the overall roadmap_input
def prepare_agent_inputs(state: RouterState) -> dict:
    roadmap_input = state["roadmap_input"]
    decision = state["decision"]
    iteration = state.get("iteration", 0) + 1

    # Adjust marketing budget based on the decision
    if decision == "Increase marketing budget":
        roadmap_input.supplied_marketing_budget += roadmap_input.total_budget*.05
    elif decision == "Decrease marketing budget":
        roadmap_input.supplied_marketing_budget -= roadmap_input.total_budget*0.05

    # Check for budget bounds here before proceeding
    if roadmap_input.supplied_marketing_budget <= 0 or roadmap_input.supplied_marketing_budget > roadmap_input.total_budget:
        # If budget becomes invalid, force a stop decision for the next router call.
        # The `decide_next_action` node will pick this up.
        return {"decision": "Stop", "roadmap_input": roadmap_input, "iteration": iteration}

    # Prepare ResourceManagerInput
    feature_items_for_resource_manager = []
    for feature in roadmap_input.features:
        feature_items_for_resource_manager.append(
            FeatureItem(
                feature_name=feature.name,
                functionality=feature.functionality,
                resources_required=roadmap_input.resources_per_feature.get(feature.name, []),
                mandatory=feature.name in roadmap_input.mandatory_features
            )
        )
    resource_budget = roadmap_input.total_budget - roadmap_input.supplied_marketing_budget
    if resource_budget < 0:
        # If resource budget becomes negative, it's impossible. Force stop.
        return {"decision": "Stop", "roadmap_input": roadmap_input, "iteration": iteration}

    resource_manager_input = ResourceManagerInput(
        featureset1=feature_items_for_resource_manager,
        total_budget=resource_budget,
        total_time_months=roadmap_input.total_time_months,
        market_stage=roadmap_input.market_stage
    )

    # # Prepare FeaturesInputList for FeatureValueAgent
    # feature_value_input = FeaturesInputList(features=roadmap_input.features)

    # # Prepare PerceivedValueInput
    # perceived_value_input = PerceivedValueInput(
    #     features=roadmap_input.features,
    #     supplied_marketing_budget=roadmap_input.supplied_marketing_budget,
    #     current_market=roadmap_input.current_market,
    #     market_stage=roadmap_input.market_stage
    #)
    return {
        "resource_manager_input": resource_manager_input,
        "roadmap_input": roadmap_input, # Update the roadmap_input in state
        "iteration": iteration # Increment iteration after preparing inputs
    }
@traceable
def query_feature_importance(state: CompetitionState) -> dict:
    feature_importance_input = state["feature_list"]
    if not feature_importance_input:
        return {"feature_importance_raw_output": None}
    current_market = state["current_market"]
    index = state["index"]
    feature_in = feature_importance_input[index]
    if not feature_in:
        return {"feature_importance_raw_output": None}

    competitive_feature_input = FeatureImportanceInput(
        feature_name = feature_in.name,
        market_name = current_market,
        functionality_name = feature_in.functionality
    )
    agent_input_message = HumanMessage(content=(
        f"Please evaluate the feature importance on the following details:\n"
        f"{competitive_feature_input.model_dump_json(indent=2)}"
    ))
    raw_agent_output = competition_agent.invoke({"messages": [agent_input_message]})
    final_message_content = ""
    if 'messages' in raw_agent_output and len(raw_agent_output['messages']) > 0:
        final_message = raw_agent_output['messages'][-1]
        if hasattr(final_message, 'content') and isinstance(final_message.content, str):
            final_message_content = final_message.content
    return {"feature_importance_raw_output": final_message_content}

def query_perceived_value(state: RouterState) -> dict:
    roadmap_input = state["roadmap_input"]
    perceived_value_input = None
    resource_manager_output = state["resource_manager_output"] #we have already must have created this object in the query_feature_value function
    if (resource_manager_output):
      if(resource_manager_output.selected_features):
          feature_items_for_resource_manager = []
          features_list = resource_manager_output.selected_features
          for x in features_list:
            for y in roadmap_input.features:
              if x.feature_name == y.name:
                feature_items_for_resource_manager.append(y)
                break
          residual_marketing_budget = roadmap_input.total_budget - resource_manager_output.total_estimated_cost
          perceived_value_input = PerceivedValueInput(
            features = feature_items_for_resource_manager,
            supplied_marketing_budget=residual_marketing_budget,
            current_market=roadmap_input.current_market,
            market_stage=roadmap_input.market_stage
          )

    agent_input_message = HumanMessage(content=(
        f"Please evaluate the perceived value based on the following details:\n"
        f"{perceived_value_input.model_dump_json(indent=2)}"
    ))
    raw_agent_output = perceived_value_agent.invoke({"messages": [agent_input_message]})
    final_message_content = ""
    if 'messages' in raw_agent_output and len(raw_agent_output['messages']) > 0:
        final_message = raw_agent_output['messages'][-1]
        if hasattr(final_message, 'content') and isinstance(final_message.content, str):
            final_message_content = final_message.content
    return {"perceived_value_raw_output": final_message_content}


def query_feature_value(state: RouterState) -> dict:
    roadmap_input = state["roadmap_input"]
    feature_value_input = {}
    resource_manager_output_obj = {}
    reasoning_parts = []
    #first format the resource manager raw output
    if state["resource_manager_raw_output"]:
        try:
            parsed_rm_output = json.loads(state["resource_manager_raw_output"])
            resource_manager_output_obj = ResourceManagerOutput.model_validate(parsed_rm_output)
            reasoning_parts.append(f"Resource Manager Agent Report:\nFeasible: {resource_manager_output_obj.feasible}\nReasoning: {resource_manager_output_obj.reasoning}")
        except json.JSONDecodeError as e:
            reasoning_parts.append(f"Error parsing Resource Manager output: {e}. Raw: {state['resource_manager_raw_output']}")
        except Exception as e:
            reasoning_parts.append(f"Error validating Resource Manager output: {e}. Raw: {state['resource_manager_raw_output']}")

    if (resource_manager_output_obj):
      if(resource_manager_output_obj.selected_features):
          feature_items_for_resource_manager = []
          features_list = resource_manager_output_obj.selected_features
          for feature_obj in features_list:
            for features_r in roadmap_input.features:
              if features_r.name == feature_obj.feature_name:
                feature_items_for_resource_manager.append(features_r)
                break
          feature_value_input = FeaturesInputList(features=feature_items_for_resource_manager)
    agent_input_message = HumanMessage(content=(
        f"Please evaluate the combined value of the following features based on the provided details:\n"
        f"{feature_value_input.model_dump_json(indent=2)}"
    ))
    raw_agent_output = feature_value_agent.invoke({"messages": [agent_input_message]})
    final_message_content = ""
    if 'messages' in raw_agent_output and len(raw_agent_output['messages']) > 0:
        final_message = raw_agent_output['messages'][-1]
        if hasattr(final_message, 'content') and isinstance(final_message.content, str):
            final_message_content = final_message.content
    return {"feature_value_raw_output": final_message_content, "feature_value_input": feature_value_input, "resource_manager_output":resource_manager_output_obj, "error_reasons":state.get("error_reasons", "/n").join(reasoning_parts)}


def query_resource_manager(state: RouterState) -> dict:
    resource_manager_input = state["resource_manager_input"]
    agent_input_message = HumanMessage(content=(
        f"Please evaluate the following feature completion scenario based on the provided details:\n"
        f"{resource_manager_input.model_dump_json(indent=2)}"
    ))
    raw_agent_output = resource_manager_agent.invoke({"messages": [agent_input_message]})
    final_message_content = ""
    if 'messages' in raw_agent_output and len(raw_agent_output['messages']) > 0:
        final_message = raw_agent_output['messages'][-1]
        if hasattr(final_message, 'content') and isinstance(final_message.content, str):
            final_message_content = final_message.content
    return {"resource_manager_raw_output": final_message_content}

def route_by_status(state: CompetitionState) -> Literal["competition_agent", "end"]:
    featurelist = state["feature_list"]
    index = state["index"]
    if index >= len(featurelist):
        return "end_node"
    else:
        return "competition_agent"

def synthesize_competitor_result(state: CompetitionState) -> dict:
    feature_value_output_obj = None
    reasoning_parts = []
    # Parse ResourceManagerOutput
    if state["feature_importance_raw_output"]:
        try:
            parsed_rm_output = json.loads(state["feature_importance_raw_output"])
            feature_value_output_obj = FeatureValue.model_validate(parsed_rm_output)
            reasoning_parts.append(f"FeatureImportanceOutput:\nImportance score: {feature_value_output_obj.overall_value_score}\nReasoning: {feature_value_output_obj.value_explanation}")
            if not feature_value_output_obj:
                reasoning_parts.append("FeatureImportanceOutput was not generated")
        except json.JSONDecodeError as e:
            reasoning_parts.append(f"Error parsing Feature Importance output: {e}. Raw: {state['feature_importance_raw_output']}")
        except Exception as e:
            reasoning_parts.append(f"Error validating Resource Manager output: {e}. Raw: {state['feature_importance_raw_output']}")

    feature_out = state.get("feature_importance_output",[])
    feature_out.append(feature_value_output_obj)
    index_out = state.get("index",0)
    index_out += 1
    error_reason = state.get("error_reason", "\n")
    error_reason.join(reasoning_parts)
    return {"feature_importance_output":feature_out, "index":index_out, "error_reasons":error_reason}


def synthesize_results(state: RouterState) -> dict:
    resource_manager_output_obj = None
    total_feature_value_output_obj = None
    perceived_value_output_obj = None
    roadmap_input = state["roadmap_input"]
    prev_roadmap_output = state["intermediate_roadmap_output"]
    resource_manager_output = state["resource_manager_output"]
    feature_importance_scores = state["all_features_importance"]
    feasible_overall = False
    if (resource_manager_output):
      feasible_overall = resource_manager_output.feasible
    #first store the previous total value
    previous_total_value = 0
    if (prev_roadmap_output):
        previous_total_value = prev_roadmap_output.total_value
    reasoning_parts = []

    # Parse ResourceManagerOutput
    if resource_manager_output:
      selected_features = resource_manager_output.selected_features
    # add the feature importance scores to the output

    # Parse TotalFeatureValueOutput
    if state["feature_value_raw_output"]:
        try:
            parsed_fv_output = json.loads(state["feature_value_raw_output"])
            total_feature_value_output_obj = TotalFeatureValueOutput.model_validate(parsed_fv_output)
            reasoning_parts.append(f"\nFeature Value Agent Report:\nTotal Value Score: {total_feature_value_output_obj.total_value_score:.2f}\nReasoning: {total_feature_value_output_obj.reasoning}")
        except json.JSONDecodeError as e:
            reasoning_parts.append(f"Error parsing Feature Value output: {e}. Raw: {state['feature_value_raw_output']}")
        except Exception as e:
            reasoning_parts.append(f"Error validating Feature Value output: {e}. Raw: {state['feature_value_raw_output']}")

    # Parse PerceivedValueOutput
    if state["perceived_value_raw_output"]:
        try:
            parsed_pv_output = json.loads(state["perceived_value_raw_output"])
            perceived_value_output_obj = PerceivedValueOutput.model_validate(parsed_pv_output)
            reasoning_parts.append(f"\nPerceived Value Agent Report:\nTotal Perceived Value: {perceived_value_output_obj.total_perceived_value:.2f}\nReasoning: {perceived_value_output_obj.reasoning}")
        except json.JSONDecodeError as e:
            reasoning_parts.append(f"Error parsing Perceived Value output: {e}. Raw: {state['perceived_value_raw_output']}")
        except Exception as e:
            reasoning_parts.append(f"Error validating Perceived Value output: {e}. Raw: {state['perceived_value_raw_output']}")
    total_value = 0.0
    if total_feature_value_output_obj:
        total_value += total_feature_value_output_obj.total_value_score
    if perceived_value_output_obj:
        #individual_perceived_values = perceived_value_output_obj.individual_feature_values
        total_value += perceived_value_output_obj.total_perceived_value
        #if we have a perceived_value_output_object, lets also fill the scheduled_features list with the values
        if(selected_features):
          for selected_feature in selected_features:
            for feature_importance_score in feature_importance_scores:
              if selected_feature.feature_name == feature_importance_score.feature_name:
                selected_feature.value = feature_importance_score.overall_value_score
                total_value += feature_importance_score.overall_value_score
                perceived_value_output_obj.total_perceived_value +=feature_importance_score.overall_value_score
                reasoning_parts.append(f"\nFeature Importance Score: {feature_importance_score.overall_value_score:.2f}\nReasoning: {feature_importance_score.value_explanation}")
                break
    intermediate_roadmap_output = RoadmapOutput(
        feasible=feasible_overall,
        total_value=total_value,
        Final_marketing_budget_usd = perceived_value_output_obj.adjusted_marketing_budget_usd,
        selected_features=selected_features,
        resource_manager_output=resource_manager_output_obj,
        total_feature_value_output=total_feature_value_output_obj,
        perceived_value_output=perceived_value_output_obj,
        summary_reasoning="\n".join(reasoning_parts) # Initial summary, LLM will refine later
    )

    return {
        "intermediate_roadmap_output": intermediate_roadmap_output,
        "prev_total_value": previous_total_value,
        "prev_decision": state["decision"]
    }

def synthesize_final_results(state: RouterState) -> dict:
    # This node will be called when the decide_next_action determines to stop.
    # It should use the latest intermediate_roadmap_output and other state info.
    intermediate_roadmap_output = state["intermediate_roadmap_output"]
    roadmap_input = state["roadmap_input"]
    final_roadmap_output = None
    reasoning_parts = []
    if(intermediate_roadmap_output):
      if intermediate_roadmap_output.resource_manager_output:
          reasoning_parts.append(f"Resource Manager Agent Report:\nFeasible: {intermediate_roadmap_output.resource_manager_output.feasible}\nReasoning: {intermediate_roadmap_output.resource_manager_output.reasoning}")
      if intermediate_roadmap_output.total_feature_value_output:
          reasoning_parts.append(f"\nFeature Value Agent Report:\nTotal Value Score: {intermediate_roadmap_output.total_feature_value_output.total_value_score:.2f}\nReasoning: {intermediate_roadmap_output.total_feature_value_output.reasoning}")
      if intermediate_roadmap_output.perceived_value_output:
          reasoning_parts.append(f"\nPerceived Value Agent Report:\nTotal Perceived Value: {intermediate_roadmap_output.perceived_value_output.total_perceived_value:.2f}\nReasoning: {intermediate_roadmap_output.perceived_value_output.reasoning}")
      #Also add error reasons from state in
      if state.get("error_reasons"):
          reasoning_parts.append(f"\nError Reasons from Feature Value porcessing:\n{state.get('error_reasons')}")

      synthesis_prompt_content = f"""Synthesize the following reports from different agents to provide a comprehensive roadmap evaluation.
      Focus on overall feasibility, key timelines from resource management, total_value, Final marketing budget usd, total feature value, and perceived value.
      Clearly state the overall feasibility.
      Reports:
      {"\n\n".join(reasoning_parts)}
      """
      synthesis_response = llm.invoke([
          {"role": "system", "content": "You are an expert roadmap evaluator. Summarize the provided agent reports into a coherent and actionable roadmap evaluation."},
          {"role": "user", "content": synthesis_prompt_content}
      ])

      final_roadmap_output = RoadmapOutput(
        feasible=intermediate_roadmap_output.feasible,
        total_value=intermediate_roadmap_output.total_value,
        Final_marketing_budget_usd=intermediate_roadmap_output.Final_marketing_budget_usd,
        selected_features=intermediate_roadmap_output.selected_features,
        resource_manager_output=intermediate_roadmap_output.resource_manager_output,
        total_feature_value_output=intermediate_roadmap_output.total_feature_value_output,
        perceived_value_output=intermediate_roadmap_output.perceived_value_output,
        summary_reasoning=synthesis_response.content)

    return {"final_roadmap_output": final_roadmap_output}
```
